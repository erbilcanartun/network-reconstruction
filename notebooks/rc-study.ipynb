```python
import numpy as np
import matplotlib.pyplot as plt

# Toy data: Mackey-Glass (chaotic series)
def mackey_glass(t, tau=17, n=10):
    # Simple numerical integration for demo
    x = np.zeros(len(t))
    x[0] = 1.2
    for i in range(1, len(t)):
        x[i] = x[i-1] + (0.2 * x[i-1] * (0.9 - x[i-tau]) / (1 + x[i-tau]**10) - 0.1 * x[i-1]) * (t[1] - t[0])
    return x

T = 1000  # Time steps
t = np.linspace(0, T, T)
u = mackey_glass(t)  # Input series (predict next value, so y = u[1:] shifted)
train_len = 800
u_train, u_test = u[:train_len], u[train_len:]
y_train = u_train[1:]  # Target: next step
```
