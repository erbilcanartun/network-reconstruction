```python
"""
Paper-faithful Spiking Liquid State Machine for closed-loop prediction of
the Rulkov map, following:

  Rosati, Gambosi, Toschi, Duggento (2025),
  "Spiking Reservoir Computing Architectures for Model-based Epileptic
   Brain State Recognition", IEEE EMBC.

The paper applies an LSM to an epileptic-parameter regression task. Here we
reuse its architecture (LIF reservoir, population coding, unsupervised STDP
pre-exposure, frozen reservoir + linear readout) for a *different* task:
closed-loop (autoregressive) prediction of the Rulkov map's fast variable.

Design decisions and fixed parameters are justified in the block comments.

------------------------------------------------------------
PRIORITY MAP FOR THE SEARCH
------------------------------------------------------------
HIGH-impact (searched densely):
  - spectral_radius     : edge-of-chaos control
  - tau_filt            : readout-feature time scale; must match Rulkov's
                          fast-spike/slow-burst time scale
  - stdp_exposure_s     : the paper's central knob; paper range 0.25..10 s
  - regression_alpha    : ridge strength; controls closed-loop stability
  - N_liquid            : paper finds sweet spot ~100, STDP helps small nets

MEDIUM-impact (searched lightly):
  - w_input             : whether input actually drives the reservoir
  - sigma_enc           : width of population-coding tuning curves
  - conn_prob           : reservoir sparsity

LOW-impact / FIXED (pinned to paper or sensible defaults):
  - LIF membrane params  (all from Table III)
  - STDP time constants  (tau_+ = tau_- = 20 ms)
  - STDP amplitudes      (A_+=10, A_-=12 in paper units; we keep A_- > A_+)
  - exc_frac             (80/20 Dale's principle, biological default)
  - dt, step_duration    (practical compromise; see note below)
------------------------------------------------------------
"""

import numpy as np
import matplotlib.pyplot as plt
import itertools
import time
from numba import njit
from sklearn.linear_model import Ridge

np.random.seed(42)

# ==========================================================
# USER SETTINGS
# ==========================================================
DATA_PATH   = "../../../data/chaotic_data/rulkov_map.csv"
COLUMN_IDX  = 0

train_len   = 4000
test_start  = 4000
test_len    = 1000
seed        = 42

# Readout feature mode. Quadratic (S, S**2) adds symmetry-breaking nonlinearity
# that helps closed-loop rollout on spiking features. Kept as a fixed design
# choice, not a search axis, because switching it re-scales the feature space
# and makes ridge_alpha comparisons meaningless across trials.
READOUT_FEATURES = "quadratic"
TRAIN_NOISE_STD  = 1e-4       # Jaeger-style regularization against closed-loop drift

# Random-search budget (full grid is too large for exhaustive search)
N_TRIALS = 300

# Separation-property guardrail
USE_SEPARATION_CHECK = True
SEP_RATIO_MIN = 0.25          # reject if plastic/initial distance ratio collapses
SEP_RATIO_MAX = 4.0           # reject runaway potentiation


# ==========================================================
# SCALERS
# ==========================================================
def fit_scaler(x, method="minmax11"):
    x = np.asarray(x).reshape(-1, 1)
    if method == "minmax11":
        xmin = x.min(axis=0)
        xmax = x.max(axis=0)
        scale = xmax - xmin
        scale[scale == 0] = 1.0
        return {"method": method, "xmin": xmin, "xmax": xmax, "scale": scale}
    raise ValueError(method)

def transform_array(x, scaler):
    x = np.asarray(x).reshape(-1, 1)
    return 2.0 * (x - scaler["xmin"]) / scaler["scale"] - 1.0

def inverse_transform_array(x, scaler):
    x = np.asarray(x).reshape(-1, 1)
    return 0.5 * (x + 1.0) * scaler["scale"] + scaler["xmin"]


# ==========================================================
# LOAD DATA
# ==========================================================
dataset = np.loadtxt(DATA_PATH, delimiter=",")
if dataset.ndim == 2:
    dataset = dataset[:, COLUMN_IDX]
dataset = dataset.astype(np.float64)

fig, ax = plt.subplots(figsize=(10, 3))
ax.plot(dataset[:2000], lw=0.7)
ax.set_title("Rulkov map — first 2000 samples")
ax.set_xlabel("Step"); ax.set_ylabel("x(n)")
plt.tight_layout(); plt.show()

data  = dataset.reshape(-1, 1)
X_raw = data[:-1]
Y_raw = data[1:]

X_train_raw = X_raw[:train_len]
Y_train_raw = Y_raw[:train_len]
X_test_raw  = X_raw[test_start:test_start + test_len]
Y_test_raw  = Y_raw[test_start:test_start + test_len]


# ==========================================================
# NUMBA KERNELS
# ==========================================================
# Notes on unit conventions (consistent with paper):
#   V in mV, I_syn in pA, C_m in nF -> dV/dt ~ I/C has units mV/ms when
#   I is pA and C is nF (since pA/nF = mV/ms).
#   Recurrent weights W_rec are in pA (currents).
#   STDP A_+, A_- are in pA, so one STDP event shifts a weight by ~10 pA
#   times trace value; trace values are O(1) so changes are meaningful.

@njit(cache=True)
def _encode_population(x_scalar, mu, sigma_enc):
    """Gaussian tuning curves over scaled signal in [-1, 1]."""
    N = mu.shape[0]
    out = np.empty(N)
    inv_2s2 = 1.0 / (2.0 * sigma_enc * sigma_enc)
    for i in range(N):
        d = x_scalar - mu[i]
        out[i] = np.exp(-d * d * inv_2s2)
    return out


@njit(cache=True)
def _compute_input_current(inp_act, W_in):
    N, M = W_in.shape
    out = np.zeros(N)
    for i in range(N):
        s = 0.0
        for j in range(M):
            s += W_in[i, j] * inp_act[j]
        out[i] = s
    return out


@njit(cache=True)
def _simulate_window(
    V, I_syn, refrac, trace_pre, trace_post,
    W_rec, I_input,
    n_substeps,
    V_rest, V_reset, V_thresh,
    decay_mem, drive_scale, decay_syn,
    refrac_steps,
    stdp_enabled, decay_trace,
    A_plus, A_minus, w_min_exc, w_max_exc, w_min_inh, w_max_inh,
    neuron_sign,
):
    """
    Advance reservoir for one sample window (step_duration / dt substeps).

    Dale-respecting STDP: a synapse from an excitatory presynaptic neuron
    stays in [0, w_max_exc]; from an inhibitory presynaptic neuron in
    [-w_max_inh, 0]. This prevents STDP from silently flipping a synapse's
    sign — the paper's Dale enforcement requires this.

    Returns per-neuron spike counts in this window.
    """
    N = V.shape[0]
    spike_counts = np.zeros(N)

    for _ in range(n_substeps):
        for i in range(N):
            trace_pre[i]  *= decay_trace
            trace_post[i] *= decay_trace
            I_syn[i]      *= decay_syn

        for i in range(N):
            if refrac[i] <= 0:
                V[i] = V_rest + (V[i] - V_rest) * decay_mem \
                       + (I_syn[i] + I_input[i]) * drive_scale
            else:
                refrac[i] -= 1

        spiked = np.zeros(N, dtype=np.uint8)
        for i in range(N):
            if V[i] >= V_thresh:
                spiked[i] = 1
                spike_counts[i] += 1.0
                V[i] = V_reset
                refrac[i] = refrac_steps

        for pre in range(N):
            if spiked[pre] == 1:
                trace_pre[pre] += 1.0
                for post in range(N):
                    I_syn[post] += W_rec[post, pre]
                if stdp_enabled:
                    for post in range(N):
                        delta = A_plus * trace_post[post]
                        new_w = W_rec[post, pre] + delta
                        if neuron_sign[pre] > 0:
                            if new_w > w_max_exc: new_w = w_max_exc
                            elif new_w < w_min_exc: new_w = w_min_exc
                        else:
                            if new_w > w_max_inh: new_w = w_max_inh
                            elif new_w < w_min_inh: new_w = w_min_inh
                        W_rec[post, pre] = new_w

        for post in range(N):
            if spiked[post] == 1:
                trace_post[post] += 1.0
                if stdp_enabled:
                    for pre in range(N):
                        delta = A_minus * trace_pre[pre]
                        new_w = W_rec[post, pre] - delta
                        if neuron_sign[pre] > 0:
                            if new_w > w_max_exc: new_w = w_max_exc
                            elif new_w < w_min_exc: new_w = w_min_exc
                        else:
                            if new_w > w_max_inh: new_w = w_max_inh
                            elif new_w < w_min_inh: new_w = w_min_inh
                        W_rec[post, pre] = new_w

    return spike_counts


@njit(cache=True)
def _run_sequence(
    input_sequence,
    V, I_syn, refrac, filt, trace_pre, trace_post,
    W_rec, W_in, mu, sigma_enc,
    n_substeps,
    V_rest, V_reset, V_thresh,
    decay_mem, drive_scale, decay_syn, decay_filt,
    refrac_steps,
    stdp_enabled, decay_trace,
    A_plus, A_minus, w_min_exc, w_max_exc, w_min_inh, w_max_inh,
    neuron_sign,
):
    T = input_sequence.shape[0]
    N = V.shape[0]
    states = np.zeros((T, N))

    for t in range(T):
        x = input_sequence[t, 0]
        inp_act = _encode_population(x, mu, sigma_enc)
        I_input = _compute_input_current(inp_act, W_in)

        counts = _simulate_window(
            V, I_syn, refrac, trace_pre, trace_post,
            W_rec, I_input,
            n_substeps,
            V_rest, V_reset, V_thresh,
            decay_mem, drive_scale, decay_syn,
            refrac_steps,
            stdp_enabled, decay_trace,
            A_plus, A_minus, w_min_exc, w_max_exc, w_min_inh, w_max_inh,
            neuron_sign,
        )
        for i in range(N):
            filt[i] = filt[i] * decay_filt + counts[i]
            states[t, i] = filt[i]

    return states


@njit(cache=True)
def _run_one_step(
    x_scalar,
    V, I_syn, refrac, filt, trace_pre, trace_post,
    W_rec, W_in, mu, sigma_enc,
    n_substeps,
    V_rest, V_reset, V_thresh,
    decay_mem, drive_scale, decay_syn, decay_filt,
    refrac_steps, neuron_sign,
):
    x_arr = np.empty((1, 1)); x_arr[0, 0] = x_scalar
    states = _run_sequence(
        x_arr,
        V, I_syn, refrac, filt, trace_pre, trace_post,
        W_rec, W_in, mu, sigma_enc,
        n_substeps,
        V_rest, V_reset, V_thresh,
        decay_mem, drive_scale, decay_syn, decay_filt,
        refrac_steps,
        False, 1.0,            # STDP off during inference
        0.0, 0.0, -1e9, 1e9, -1e9, 1e9,
        neuron_sign,
    )
    return states[0].copy()


# ==========================================================
# LSM WRAPPER
# ==========================================================
class SpikingLSM:
    """
    Paper-faithful LSM.

    Fixed to paper Table III:
      V_rest=-65, V_reset=-65, V_thresh=-50, tau_m=8, tau_syn=2, tau_ref=0
      STDP: A_+=10 pA, A_-=12 pA, tau_+=tau_-=20 ms
    Searched:
      N_liquid, spectral_radius, tau_filt, sigma_enc, conn_prob, w_input
      (see param_grid below)
    """
    def __init__(
        self,
        N_liquid=100,
        N_input=50,
        # Paper-fixed LIF params
        tau_mem=8.0,
        tau_syn=2.0,
        V_rest=-65.0, V_reset=-65.0, V_thresh=-50.0,
        refrac_ms=0.0,
        # Simulation time (paper: dt=0.1 ms, T=500 ms; we use dt=1 ms for speed)
        dt=1.0,
        step_duration=50.0,
        # Readout filtering
        tau_filt=100.0,
        # Connectivity
        exc_frac=0.8,
        conn_prob=0.1,
        input_conn_prob=0.3,
        # Weight magnitude (variance ~ 1/N per Yildiz et al.)
        spectral_radius=0.95,
        w_input=30.0,            # in pA; must be large enough to drive LIF
        # Population coding
        sigma_enc=0.2,
        # STDP (paper-fixed)
        A_plus=10.0,
        A_minus=12.0,
        tau_trace=20.0,
        # STDP weight clamps. Separate for E and I to respect Dale's law.
        w_max_exc=80.0,
        w_max_inh=80.0,
        rng_seed=42,
    ):
        self.N = N_liquid
        self.N_input = N_input
        self.n_substeps = max(1, int(round(step_duration / dt)))

        self.V_rest = V_rest
        self.V_reset = V_reset
        self.V_thresh = V_thresh
        self.refrac_steps = int(round(refrac_ms / dt))

        # Exp-decay factors for integration. dt in ms, time constants in ms.
        self.decay_mem   = np.exp(-dt / tau_mem)
        self.drive_scale = 1.0 - self.decay_mem
        self.decay_syn   = np.exp(-dt / tau_syn)
        self.decay_filt  = np.exp(-step_duration / tau_filt)
        self.decay_trace = np.exp(-dt / tau_trace)

        self.mu = np.linspace(-1.0, 1.0, N_input)
        self.sigma_enc = sigma_enc

        self.A_plus = A_plus
        self.A_minus = A_minus
        self.w_min_exc = 0.0
        self.w_max_exc = w_max_exc
        self.w_min_inh = -w_max_inh
        self.w_max_inh = 0.0

        rng = np.random.RandomState(rng_seed)

        # --- Initialization per Yildiz/Jaeger/Kiebel (2012) ---
        # 1) non-negative entries, variance ~1/N
        # 2) scale so spectral radius equals target (paper says <1; we search
        #    both sides of 1 since edge-of-chaos is task-dependent)
        # 3) flip signs by Dale's principle: whole-neuron E or I, not per-synapse
        n_exc = int(exc_frac * N_liquid)
        neuron_sign = np.ones(N_liquid)
        neuron_sign[n_exc:] = -1.0
        rng.shuffle(neuron_sign)  # randomize which indices are I
        self.neuron_sign = neuron_sign.astype(np.float64)

        std_init = 1.0 / np.sqrt(N_liquid)
        W_rec = np.abs(rng.randn(N_liquid, N_liquid) * std_init)

        mask = rng.rand(N_liquid, N_liquid) < conn_prob
        np.fill_diagonal(mask, False)
        W_rec *= mask

        # Apply Dale: whole presynaptic column gets the sign of that neuron.
        for pre in range(N_liquid):
            if self.neuron_sign[pre] < 0:
                W_rec[:, pre] *= -1.0

        # Spectral rescaling (preserves signs — just a scalar multiply)
        try:
            eigvals = np.linalg.eigvals(W_rec)
            sr = float(np.max(np.abs(eigvals)))
            if sr > 1e-12:
                W_rec *= (spectral_radius / sr)
        except np.linalg.LinAlgError:
            pass

        # Convert to pA scale. std_init above was dimensionless; typical
        # LIF drive needs tens of pA to spike given (V_thresh - V_rest)=15 mV
        # and tau_m=8 ms. We multiply by a base current scale.
        W_rec *= 60.0  # pA-scale so individual synapses can meaningfully drive

        self.W_rec = W_rec.astype(np.float64)

        # Input weights (excitatory only; thalamic-like drive)
        W_in = np.zeros((N_liquid, N_input), dtype=np.float64)
        in_mask = rng.rand(N_liquid, N_input) < input_conn_prob
        W_in[in_mask] = rng.uniform(0.0, w_input, size=np.sum(in_mask))
        self.W_in = W_in

        self.reset_state()

    def reset_state(self):
        self.V = np.full(self.N, self.V_rest, dtype=np.float64)
        self.I_syn = np.zeros(self.N, dtype=np.float64)
        self.refrac = np.zeros(self.N, dtype=np.int32)
        self.filt = np.zeros(self.N, dtype=np.float64)
        self.trace_pre = np.zeros(self.N, dtype=np.float64)
        self.trace_post = np.zeros(self.N, dtype=np.float64)

    def copy_recurrent_weights(self):
        return self.W_rec.copy()

    def run_sequence(self, input_sequence, stdp_enabled=False):
        return _run_sequence(
            np.ascontiguousarray(input_sequence),
            self.V, self.I_syn, self.refrac, self.filt,
            self.trace_pre, self.trace_post,
            self.W_rec, self.W_in, self.mu, self.sigma_enc,
            self.n_substeps,
            self.V_rest, self.V_reset, self.V_thresh,
            self.decay_mem, self.drive_scale, self.decay_syn, self.decay_filt,
            self.refrac_steps,
            stdp_enabled, self.decay_trace,
            self.A_plus, self.A_minus,
            self.w_min_exc, self.w_max_exc, self.w_min_inh, self.w_max_inh,
            self.neuron_sign,
        )

    def run_one_step(self, x_scalar):
        return _run_one_step(
            float(x_scalar),
            self.V, self.I_syn, self.refrac, self.filt,
            self.trace_pre, self.trace_post,
            self.W_rec, self.W_in, self.mu, self.sigma_enc,
            self.n_substeps,
            self.V_rest, self.V_reset, self.V_thresh,
            self.decay_mem, self.drive_scale, self.decay_syn, self.decay_filt,
            self.refrac_steps, self.neuron_sign,
        )


# ==========================================================
# READOUT FEATURES
# ==========================================================
def make_features(S, mode=READOUT_FEATURES):
    if mode == "linear":
        return S
    if mode == "quadratic":
        return np.hstack([S, S**2])
    raise ValueError(mode)


# ==========================================================
# SEPARATION PROPERTY  (paper eq. 5, ratio form)
# ==========================================================
def _state_distance(Sa, Sb):
    T = min(Sa.shape[0], Sb.shape[0])
    Sa, Sb = Sa[:T], Sb[:T]
    return float(np.mean(np.linalg.norm(Sa - Sb, axis=1)))

def separation_ratio(lsm_plastic_fn_build, W_before, W_after, probe_seq,
                      noise_std=0.02, seed=0):
    """
    Paper-style SP check: compare < d'(Xu, Xv)_plastic > / < d'(Xu, Xv)_initial >
    for matched probe pairs (u, v). Ratio near 1 = SP preserved.

    We build two LSM copies with identical init seed so W_in and neuron signs
    match; only W_rec differs.
    """
    rng = np.random.RandomState(seed)
    u = probe_seq.copy()
    v = probe_seq.copy() + rng.normal(0.0, noise_std, size=probe_seq.shape)

    lsm_initial = lsm_plastic_fn_build()
    lsm_initial.W_rec = W_before.copy()
    lsm_initial.reset_state(); Sa0 = lsm_initial.run_sequence(u)
    lsm_initial.reset_state(); Sb0 = lsm_initial.run_sequence(v)
    d0 = _state_distance(Sa0, Sb0)

    lsm_plastic = lsm_plastic_fn_build()
    lsm_plastic.W_rec = W_after.copy()
    lsm_plastic.reset_state(); SaP = lsm_plastic.run_sequence(u)
    lsm_plastic.reset_state(); SbP = lsm_plastic.run_sequence(v)
    dP = _state_distance(SaP, SbP)

    if d0 < 1e-9:
        return np.nan
    return dP / d0


# ==========================================================
# EVALUATION (one parameter combo -> NRMSE on closed-loop rollout)
# ==========================================================
def evaluate_lsm(params, rng_seed=42):
    try:
        N_liquid        = params["N_liquid"]
        spectral_radius = params["spectral_radius"]
        tau_filt        = params["tau_filt"]
        sigma_enc       = params["sigma_enc"]
        conn_prob       = params["conn_prob"]
        w_input         = params["w_input"]
        stdp_exposure_s = params["stdp_exposure_s"]
        regression      = params["regression_alpha"]
        train_warmup    = params["train_warmup"]
        test_warmup     = params["test_warmup"]

        if test_warmup >= test_len:
            return np.inf, None

        scaler  = fit_scaler(X_train_raw, method="minmax11")
        X_train = transform_array(X_train_raw, scaler)
        Y_train = transform_array(Y_train_raw, scaler)
        X_test  = transform_array(X_test_raw, scaler)
        Y_test  = transform_array(Y_test_raw, scaler)

        x_lo = float(X_train.min()) - 0.05
        x_hi = float(X_train.max()) + 0.05

        def build_lsm():
            return SpikingLSM(
                N_liquid=N_liquid, N_input=50,
                spectral_radius=spectral_radius,
                tau_filt=tau_filt,
                sigma_enc=sigma_enc,
                conn_prob=conn_prob,
                w_input=w_input,
                rng_seed=rng_seed,
            )

        lsm = build_lsm()

        # ---- STDP exposure (paper: 0.25..10 s simulated time) ----
        # Each input sample occupies step_duration ms of sim time.
        step_duration_ms = lsm.n_substeps  # since dt = 1 ms
        samples_per_second = 1000.0 / step_duration_ms
        n_expose_samples = int(round(stdp_exposure_s * samples_per_second))
        n_expose_samples = min(n_expose_samples, train_len)

        W_before = lsm.copy_recurrent_weights()

        if n_expose_samples > 0:
            lsm.reset_state()
            _ = lsm.run_sequence(X_train[:n_expose_samples], stdp_enabled=True)

        W_after = lsm.copy_recurrent_weights()

        # ---- Separation-property guardrail (paper eq. 5) ----
        sep_ratio = np.nan
        if USE_SEPARATION_CHECK and n_expose_samples > 0:
            probe = X_train[:200]
            sep_ratio = separation_ratio(
                build_lsm, W_before, W_after, probe,
                noise_std=0.02, seed=rng_seed
            )
            if not np.isfinite(sep_ratio):
                return np.inf, None
            if sep_ratio < SEP_RATIO_MIN or sep_ratio > SEP_RATIO_MAX:
                return np.inf, None

        # ---- Collect frozen-reservoir training states ----
        lsm.reset_state()
        S_train = lsm.run_sequence(X_train, stdp_enabled=False)

        act = S_train.mean()
        if not np.isfinite(act) or act < 1e-6 or act > 1e6:
            return np.inf, None

        S_fit = S_train[train_warmup:]
        Y_fit = Y_train[train_warmup:, 0]

        if TRAIN_NOISE_STD > 0:
            rng_noise = np.random.RandomState(rng_seed + 123)
            S_fit = S_fit + rng_noise.normal(0.0, TRAIN_NOISE_STD,
                                             size=S_fit.shape)

        F_fit = make_features(S_fit)
        readout = Ridge(alpha=regression)
        readout.fit(F_fit, Y_fit)

        # ---- Test sync (teacher-forced warmup) ----
        lsm.reset_state()
        _ = lsm.run_sequence(X_test[:test_warmup], stdp_enabled=False)

        # ---- Closed-loop rollout ----
        pred_len = test_len - test_warmup
        Y_true_scaled = Y_test[test_warmup:test_warmup + pred_len, 0]
        Y_pred_scaled = np.zeros(pred_len)

        current_input = float(X_test[test_warmup, 0])
        for k in range(pred_len):
            state = lsm.run_one_step(current_input)
            feat = make_features(state.reshape(1, -1))
            pred = readout.predict(feat)[0]
            if not np.isfinite(pred) or abs(pred) > 1e6:
                return np.inf, None
            pred = float(np.clip(pred, x_lo, x_hi))
            Y_pred_scaled[k] = pred
            current_input = pred

        rmse = np.sqrt(np.mean((Y_true_scaled - Y_pred_scaled) ** 2))
        denom = np.std(Y_true_scaled)
        if denom == 0:
            return np.inf, None
        nrmse = rmse / denom

        return float(nrmse), {
            "scaler": scaler,
            "sep_ratio": sep_ratio,
            "W_before": W_before,
            "W_after": W_after,
            "n_expose_samples": n_expose_samples,
        }
    except Exception as e:
        print("ERROR:", e)
        return np.inf, None


# ==========================================================
# PARAMETER GRID
# ==========================================================
# Design rationale for each axis:
#
# N_liquid:         Paper sweet spot ~100; small nets benefit most from STDP.
#                   For Rulkov (richer dynamics than a single 2-param signal)
#                   we include a few larger sizes.
# spectral_radius:  Paper says rho<1 for stability, but *edge of chaos*
#                   (rho ~ 1) is standard for chaotic-signal tasks. We sweep
#                   both sides.
# tau_filt:         Rulkov has ~10-20 step bursts at alpha~4.1. tau_filt in
#                   units of step_duration*samples sets how far back the
#                   readout sees. We scan a wide range.
# sigma_enc:        Tuning-curve width. 0.1 is sharp (fewer input neurons
#                   co-active), 0.3 is broad. Too sharp = no info; too broad
#                   = no selectivity.
# conn_prob:        Sparse cortical-like connectivity.
# w_input:          In pA. LIF needs tens of pA to fire reliably.
# stdp_exposure_s:  PAPER RANGE 0.25..10 s. This is the main claim axis.
# regression_alpha: Ridge; closed-loop stability is very sensitive to this.
# train_warmup / test_warmup: short compared to train_len.
# ------------------------------------------------------------

param_grid = {
    # HIGH-priority dense grids
    "N_liquid":         [80, 120, 200, 400],
    "spectral_radius":  [0.7, 0.9, 1.0, 1.1, 1.3],
    "tau_filt":         [50.0, 100.0, 200.0, 400.0, 800.0],
    "regression_alpha": [1e-8, 1e-6, 1e-4, 1e-2, 1.0],
    "stdp_exposure_s":  [0.0, 0.25, 0.5, 1.0, 2.5, 5.0, 10.0],

    # MEDIUM-priority
    "w_input":          [15.0, 30.0, 60.0],
    "sigma_enc":        [0.12, 0.20, 0.30],
    "conn_prob":        [0.08, 0.15],

    # LOW-priority / basically pinned
    "train_warmup":     [200],
    "test_warmup":      [200],
}

keys = list(param_grid.keys())
all_combos = list(itertools.product(*[param_grid[k] for k in keys]))

rng_sampler = np.random.RandomState(seed)
n_trials = min(N_TRIALS, len(all_combos))
idx = rng_sampler.choice(len(all_combos), size=n_trials, replace=False)
combos = [all_combos[i] for i in idx]

print(f"Full grid size      : {len(all_combos)}")
print(f"Random-sampled size : {len(combos)}")


# ==========================================================
# WARM-UP COMPILE
# ==========================================================
print("Compiling Numba kernels...")
t0 = time.time()
_dummy = SpikingLSM(N_liquid=40, rng_seed=0)
_dummy_in = np.zeros((5, 1))
_dummy.run_sequence(_dummy_in, stdp_enabled=False)
_dummy.run_sequence(_dummy_in, stdp_enabled=True)
_dummy.run_one_step(0.0)
print(f"Compilation done in {time.time() - t0:.2f}s")


# ==========================================================
# SEARCH LOOP
# ==========================================================
results = []
best_nrmse = np.inf
best_params = None
best_aux = None

t_search = time.time()
for i, combo in enumerate(combos):
    params = dict(zip(keys, combo))
    nrmse, aux = evaluate_lsm(params, rng_seed=seed)

    row = params.copy()
    row["nrmse"] = nrmse
    row["sep_ratio"] = aux["sep_ratio"] if aux else np.nan
    results.append(row)

    if nrmse < best_nrmse:
        best_nrmse = nrmse
        best_params = params.copy()
        best_aux = aux

    if (i + 1) % 10 == 0 or (i + 1) == len(combos):
        el = time.time() - t_search
        eta = el / (i + 1) * (len(combos) - (i + 1))
        print(f"[{i+1:3d}/{len(combos)}] "
              f"elapsed={el:.1f}s ETA={eta:.1f}s best_NRMSE={best_nrmse:.6f}")

total_time = time.time() - t_search


# ==========================================================
# REPORT
# ==========================================================
print("\n" + "=" * 80)
print("LSM RANDOM SEARCH COMPLETE")
print(f"Train : [0:{train_len}]   Test : [{test_start}:{test_start+test_len}]")
print(f"Total time    : {total_time:.1f}s")
print(f"Best NRMSE    : {best_nrmse:.6f}")
print("Best parameters:")
for k, v in best_params.items():
    print(f"  {k:20s} = {v}")
print("=" * 80)

results_sorted = sorted(results, key=lambda r: r["nrmse"])
print("\nTop 10 configurations:")
header = (f"{'#':>3}  {'N':>4}  {'SR':>5}  {'tau_f':>6}  {'sig':>5}  {'cp':>5}  "
          f"{'w_in':>6}  {'stdp_s':>7}  {'reg':>9}  {'sep':>6}  {'NRMSE':>9}")
print(header); print("-" * len(header))
for rk, r in enumerate(results_sorted[:10], 1):
    sep = r.get("sep_ratio", np.nan)
    sep_str = f"{sep:6.2f}" if np.isfinite(sep) else "  n/a "
    print(f"{rk:3d}  {r['N_liquid']:4d}  {r['spectral_radius']:5.2f}  "
          f"{r['tau_filt']:6.1f}  {r['sigma_enc']:5.2f}  "
          f"{r['conn_prob']:5.2f}  {r['w_input']:6.1f}  "
          f"{r['stdp_exposure_s']:7.2f}  {r['regression_alpha']:9.1e}  "
          f"{sep_str}  {r['nrmse']:9.6f}")


# ==========================================================
# REBUILD BEST MODEL AND PRODUCE FINAL PLOT
# ==========================================================
bp = best_params

scaler = fit_scaler(X_train_raw, method="minmax11")
X_train_b = transform_array(X_train_raw, scaler)
Y_train_b = transform_array(Y_train_raw, scaler)
X_test_b  = transform_array(X_test_raw, scaler)
Y_test_b  = transform_array(Y_test_raw, scaler)

x_lo = float(X_train_b.min()) - 0.05
x_hi = float(X_train_b.max()) + 0.05

lsm = SpikingLSM(
    N_liquid=bp["N_liquid"], N_input=50,
    spectral_radius=bp["spectral_radius"],
    tau_filt=bp["tau_filt"],
    sigma_enc=bp["sigma_enc"],
    conn_prob=bp["conn_prob"],
    w_input=bp["w_input"],
    rng_seed=seed,
)
step_duration_ms = lsm.n_substeps  # dt = 1 ms
samples_per_second = 1000.0 / step_duration_ms
n_expose_samples = min(int(round(bp["stdp_exposure_s"] * samples_per_second)),
                       train_len)

if n_expose_samples > 0:
    lsm.reset_state()
    _ = lsm.run_sequence(X_train_b[:n_expose_samples], stdp_enabled=True)

lsm.reset_state()
S_train = lsm.run_sequence(X_train_b, stdp_enabled=False)

S_fit = S_train[bp["train_warmup"]:]
Y_fit = Y_train_b[bp["train_warmup"]:, 0]
if TRAIN_NOISE_STD > 0:
    rng_noise = np.random.RandomState(seed + 123)
    S_fit = S_fit + rng_noise.normal(0.0, TRAIN_NOISE_STD, size=S_fit.shape)

F_fit = make_features(S_fit)
readout = Ridge(alpha=bp["regression_alpha"])
readout.fit(F_fit, Y_fit)

lsm.reset_state()
_ = lsm.run_sequence(X_test_b[:bp["test_warmup"]], stdp_enabled=False)

pred_len = test_len - bp["test_warmup"]
Y_pred_scaled = np.zeros(pred_len)
current_input = float(X_test_b[bp["test_warmup"], 0])

for k in range(pred_len):
    state = lsm.run_one_step(current_input)
    feat = make_features(state.reshape(1, -1))
    pred = float(np.clip(readout.predict(feat)[0], x_lo, x_hi))
    Y_pred_scaled[k] = pred
    current_input = pred

Y_pred = inverse_transform_array(Y_pred_scaled.reshape(-1, 1), scaler).ravel()
Y_true = inverse_transform_array(
    Y_test_b[bp["test_warmup"]:bp["test_warmup"] + pred_len], scaler
).ravel()

mse  = float(np.mean((Y_true - Y_pred) ** 2))
rmse = float(np.sqrt(mse))
nrmse = rmse / float(np.std(Y_true))

print("\nFinal best-model metrics (original scale):")
print(f"  MSE   = {mse:.8f}")
print(f"  RMSE  = {rmse:.8f}")
print(f"  NRMSE = {nrmse:.8f}")
print(f"  Mean reservoir activity = {S_train.mean():.4f}")


# ==========================================================
# CLOSED-LOOP PREDICTION PLOT
# ==========================================================
fig, axes = plt.subplots(2, 1, figsize=(14, 7), sharex=False)

axes[0].plot(np.arange(test_len), Y_test_raw[:, 0],
             c="green", lw=1.2, label="Ground truth")
axes[0].axvline(bp["test_warmup"], ls="--", c="k", lw=1.0, label="Warmup end")
axes[0].plot(np.arange(bp["test_warmup"], bp["test_warmup"] + pred_len),
             Y_pred, ls="--", c="red", lw=1.2, label="LSM closed-loop")
axes[0].set_ylabel("x(n)")
axes[0].set_title(
    f"Spiking LSM closed-loop Rulkov prediction | NRMSE={nrmse:.4f} | "
    f"N={bp['N_liquid']}, SR={bp['spectral_radius']}, "
    f"tau_f={bp['tau_filt']}, STDP={bp['stdp_exposure_s']}s, "
    f"reg={bp['regression_alpha']:.0e}"
)
axes[0].legend(loc="upper right")

# zoomed-in view of the first 200 closed-loop steps
zw = min(200, pred_len)
axes[1].plot(np.arange(bp["test_warmup"], bp["test_warmup"] + zw),
             Y_true[:zw], c="green", lw=1.3, label="Ground truth")
axes[1].plot(np.arange(bp["test_warmup"], bp["test_warmup"] + zw),
             Y_pred[:zw], ls="--", c="red", lw=1.3, label="Prediction")
axes[1].set_xlabel("Step"); axes[1].set_ylabel("x(n)")
axes[1].set_title(f"Zoom: first {zw} closed-loop steps")
axes[1].legend(loc="upper right")

plt.tight_layout()
plt.savefig("/mnt/user-data/outputs/rulkov_closedloop_prediction.png", dpi=120)
plt.show()


# ==========================================================
# STDP EFFECT VISUALIZATION
# ==========================================================
if best_aux is not None and best_aux["n_expose_samples"] > 0:
    Wb = best_aux["W_before"]
    Wa = best_aux["W_after"]

    fig, axes = plt.subplots(1, 3, figsize=(14, 4))
    vmax = max(np.abs(Wb).max(), np.abs(Wa).max())
    im0 = axes[0].imshow(Wb, aspect="auto", cmap="RdBu_r",
                         vmin=-vmax, vmax=vmax)
    axes[0].set_title("W_rec before STDP")
    plt.colorbar(im0, ax=axes[0], fraction=0.046)

    im1 = axes[1].imshow(Wa, aspect="auto", cmap="RdBu_r",
                         vmin=-vmax, vmax=vmax)
    axes[1].set_title(f"W_rec after STDP ({bp['stdp_exposure_s']}s)")
    plt.colorbar(im1, ax=axes[1], fraction=0.046)

    diff = Wa - Wb
    dmax = np.abs(diff).max() + 1e-12
    im2 = axes[2].imshow(diff, aspect="auto", cmap="RdBu_r",
                         vmin=-dmax, vmax=dmax)
    axes[2].set_title("Delta W_rec")
    plt.colorbar(im2, ax=axes[2], fraction=0.046)

    plt.tight_layout()
    plt.savefig("/mnt/user-data/outputs/stdp_weight_change.png", dpi=120)
    plt.show()

print("\nDone.")
```
