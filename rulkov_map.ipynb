```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rulkov_map(beta, sigma, mu, x0, y0, iterations):
    x = np.zeros(iterations)
    y = np.zeros(iterations)
    x[0], y[0] = x0, y0
    
    for n in range(iterations-1):
        x[n+1] = beta / (1 + x[n]**2) + y[n]
        y[n+1] = y[n] - mu*x[n] - sigma
    
    return x, y
```

```python
# Parameters for chaotic behavior
beta = 4.1
sigma = 0.001
mu = 0.001
x0 = 0.1
y0 = 0.1
iterations = 10000

# Generate the time series
x, y = rulkov_map(beta, sigma, mu, x0, y0, iterations)

# Plotting
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.plot(x[2000:3000], label='Membrane Potential $x_n$')
plt.title('Rulkov Map Chaotic Time Series')
plt.xlabel('Time Steps')
plt.ylabel('Membrane Potential')
plt.legend()

plt.subplot(1, 2, 2)
#plt.plot(x, y, '.', markersize=1)
plt.title('Phase Space')
#plt.xlabel('$x_n$')
#plt.ylabel('$y_n$')

plt.plot(x[:-1], x[1:], '.', markersize=1)
plt.xlabel('$x_n$')
plt.ylabel('$x_{n+1}$')

plt.tight_layout()
plt.show()
```
